# Methods
1. Deleting the columns with missing values
2. Deleting the row with missing data
3. Fill the missing values - imputation
       (i). Numerical Data - use Mean
       (ii). Categorical Data - use Mode or new Catogory
4. Advanced Imputaion - interpolation

```python
import pandas as pd
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
```

# Deleting the column with missing data

```python
 df1 = df.copy()
```

```python
df1.isna().sum()
```

```python
df1.drop("last_review", axis = 1, inplace = True)
```

```python
df1.isna().sum()
```

# Imputation

```python
df3 = df.copy()
```

```python
df3.head()
```

```python
mean_value = df3["reviews_per_month"].mean()
df3['reviews_per_month'].fillna(mean_value, inplace=True)
```
